# **Part 1:** Theme Park Data Analysis

```python
import numpy as np

# Extract data
filename = "droptower_vdata.txt"
data_file = open(filename)
time, velocity = np.loadtxt(data_file, unpack = True)

print(time)
print(velocity)
```

```python
# Compute Position
initial_position = 30
position = initial_position + np.cumsum(velocity)

print(position.shape)
print(position)
```

```python
# Compute acceleration
acceleration = np.diff(velocity)

print(acceleration.shape)
print(acceleration)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(30,10))

# Plot position data
plt.subplot(1,3,1)
plt.title("Ride Position", fontsize=26)
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.plot(time,position)

# Plot velocity data
plt.subplot(1,3,2)
plt.title("Ride Velocity", fontsize=26)
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.plot(time,velocity)

# Plot acceleration data
plt.subplot(1,3,3)
plt.title("Ride Acceleration", fontsize=26)
plt.xlabel("Time (s)")
plt.ylabel("Acceleration (m/s^2)")
plt.plot(time[1:],acceleration)
```

# **Part 2:** Drop Tower Continued

```python
positive = acceleration[acceleration > 0]
average = np.mean(positive)
print("The average upward acceleration of the cab is", round(average,2), "m/s")
```
